````python
# Step 1: Review your code from the previous exercise and identify areas 
# where you can apply comprehension to simplify and streamline your code.
# Apply it to initialize inventory, updating the inventory and printing the updated inventory

products = ['t-shirt','mug','hat','book','keychain']

# 1 (Old code)
# def initialize_inventory(products):
#    inventory = {}
#    for element in products:
#        inventory[element] = int(input(f'Please enter your {element} stock:'))
#    return inventory

# inventory=initialize_inventory(products)

# 1/a. comprehension to initialize inventory

def initialize_inventory(products):
    inventory = {element:int(input(f'Enter the quantity of {element} available:')) for element in products}
    return inventory

# 1/b. comprehension to update inventory

def update_inventory(customer_orders,inventory):
    inventory[key]={key-1 for key in customer_orders}
    inventory
    return inventory

# 1/c. comprehension for printing the updated inventory

def print_updated_inventory(inventory):
    print('Updated_inventory:')
    [print(f'{key}:{value}') for key,value in inventory.items()]


# 2. Modify the function get_customer_orders so it prompts the user to enter the number 
# of customer orders and gathers the product names using a loop and user input. Use comprehension.

# comprehension for num_customer_orders will gather how many times a product will be prompted 
# and collected for the new customer_orders set

def get_customer_orders():
    num_customer_orders= int(input('Enter the number of customer orders:'))
    customer_orders={input('Enter name of product to order:') for i in range(num_customer_orders)}
    return customer_orders


# 3. Add a new function to calculate the total price of the customer order. 
# For each product in customer_orders, prompt the user to enter the price of that product. 
# Use comprehension to calculate the total price. Note: assume that the user can only have 1 unit of each product.


def customer_order_price(customer_orders):
    total_price=sum(float(input(f'Enter the price of the {product}: ')) for product in customer_orders)
    return total_price


# Function to calculate order stat
def calculate_order_statistics(customer_orders,products):
    total_products_ordered=len(customer_orders)
    per_product_ordered=(total_products_ordered/len(products))*100
    return total_products_ordered, per_product_ordered

# Function to print order_statistics:
def print_order_statistics(order_statistics):
    print('Order statistics: ')
    print('Total products ordered: ', order_statistics[0])
    print('Percentage of unique products ordered:', order_statistics[1],'%')


# 4. Modify the update_inventory function to remove the product from the inventory if 
# its quantity becomes zero after fulfilling the customer orders. 
# Use comprehension to filter out the products with a quantity of zero from the inventory.

def update_inventory(customer_orders,inventory):
    inventory={key:inventory[key]-1 if key in customer_orders and inventory else inventory[key] for key in inventory}
    inventory={key:value for key,value in inventory.items() if value>0}
    return inventory

# 5. Print total price of customer_orders

print('\nTotal price: ', total_price)
````

````python
# Call all functions

products = ['t-shirt','mug','hat','book','keychain']
inventory=initialize_inventory(products)
customer_orders=get_customer_orders()

order_statistics=calculate_order_statistics(customer_orders,products)
print_order_statistics(order_statistics)

inventory=update_inventory(customer_orders,inventory)
print_updated_inventory(inventory)

total_price=customer_order_price(customer_orders)
print('\nTotal price: ', total_price)
````
